```python
import numpy as np
from scipy.signal import convolve2d

def compute_sobel_edges(matrix):
    # Prepare the kernels
    weight = np.matrix([1, 2, 1])
    deri = np.matrix([-1, 0, 1])
    Kx = weight.T * deri
    Ky = -1*deri.T * weight
    print(Kx)
    print(Ky)
    # Apply the Sobel operator
    Gx = convolve2d(matrix, Kx, mode="same", boundary="symm")
    print(Gx)
    Gy = convolve2d(matrix, Ky, mode="same", boundary="symm")
    print(Gy)
    # G = np.sqrt(Gx**2 + Gy**2)
    # # or using the absolute values
    G = np.abs(Gx) + np.abs(Gy)
    return G
```

```python
M = np.array(
    [[ 4,  5,  2,  4,  8,  5,],
    [ 8, 11,  8, 11, 12, 13,],
    [12, 14, 17, 16, 16, 20,],
    [21, 19, 21, 23, 25, 25,],
    [28, 26, 29, 30, 30, 32,],
    [34, 32, 32, 33, 38, 36,]]
)
G = compute_sobel_edges(matrix=M)
print(G)
```
