```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_function_and_derivative(x, func, derivative):
    
    # setting the axes at the centre
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    # plot the function
    ax.plot(x,func(x), color="#d35400", linewidth=3, label=func.__name__)
    ax.plot(x,derivative(x), color="#1abd15", linewidth=3, label=derivative.__name__)
    
    ax.legend(loc="upper left", frameon=False)
    plt.show()
```

### 1.1 Exponential

```python
def exponential(x):                                        
    return np.exp(x)

def exponential_derivative(x):
    return np.exp(x)
```

```python
x = np.arange(-5, 5, 0.01)
plot_function_and_derivative(x, exponential, exponential_derivative)
```

```python
# o hm trung tm

def func(x):
    return x**2 - 2

def gradient(f, x, epsilon):     
    return (f(x + epsilon/2) - f(x - epsilon/2)) / epsilon

print(f'{gradient(f=func, x=1.0, epsilon=1.0e-10)}')
```

```python
# Newton method to find a solution of a function

def compute_square_root2(num_loops):    
    x_n = 0.01
    epsilon=1.0e-10
    print(x_n)
    
    for _ in range(num_loops):
        x_np1 = x_n - (func(x_n) / gradient(func, x_n, epsilon))
        x_n = x_np1
        print(x_np1)
        
    return x_np1

print(compute_square_root2(num_loops=15))
```

```python
import tensorflow as tf

def func(x):
    return x**2 - 2

def gradient(f, x):
    # create variable
    x = tf.Variable(x)

    with tf.GradientTape() as tape:    
        # construct functions
        f_x = f(x)

    # compute derivative
    df_dx = tape.gradient(f_x, x)
    return df_dx.numpy()
    
print(gradient(func, 1.0))
```

```python
def approximate_sin(x, n):
    result = 0
    for i in range(n+1):
        result += (-1)**i * x**(2*i+1)/frac(2*i + 1)
    return result
```
